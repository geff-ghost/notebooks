# Key Questions to Explore  

## Q1:What is the average daily phone usage?

```python
df.columns
```

```python
# all statistical information about daily phone usage
df['daily_screen_time_hours'].describe()
```

```python
avg_screen_time = df['daily_screen_time_hours'].mean()

print(f"Average Screen Time: {avg_screen_time:.2f} hours.")
```

I have extracted these characteristics from the data set to explore them and find out the behavior of addicted people with these behaviors.

```python
areas_of_intrests = ['user_id', 'age', 'gender', 'social_media_hours', 'notifications_per_day', 'sleep_hours', 'stress_level', 'app_opens_per_day', 'weekend_screen_time']
```

```python
people_behavior = df[df['daily_screen_time_hours'] > avg_screen_time][areas_of_intrests]
people_behavior.reset_index(drop=True, inplace=True)
```

I want to group them in different age groups, but first I must check the highest and the lowest ages with in this data frame

```python
people_behavior['age'].agg(['max', 'min'])
```

```python
labels = ['18-22', '23-27', '28-31', '32-35']
bins = [18, 23, 28, 32, 35]

people_behavior['age_group'] = pd.cut(people_behavior['age'], labels=labels, bins=bins)

people_behavior.columns
```

```python
ax = sns.countplot(people_behavior, x='age_group')
for container in ax.containers:
    ax.bar_label(container)

plt.xticks(rotation=45)
plt.tight_layout()
```

```python
age_group_count = people_behavior.groupby('age_group')['age'].count()
```

```python
age_group_count.plot(kind='bar')
plt.title('Age Group Count of People with Interested behavior')
plt.xlabel('Count')
plt.ylabel('Age Groups')
plt.tight_layout()
```

I can explore the data of consumers whose daily screen time is above average and see their addiction level

```python
addiction_level_above_avr_screen_time = df[df['daily_screen_time_hours'] > avg_screen_time]['addiction_level']

ax = addiction_level_above_avr_screen_time.value_counts().plot(kind='bar')
for container in ax.containers:
    ax.bar_label(container)
plt.xlabel('Addiction Level')
plt.ylabel('Count')
plt.title('Count of Addiction Level above Average Daily Screen Hours')
plt.xticks(rotation=45)
plt.tight_layout()
```

```python
areas_of_intrests
```

```python
gender_count_people_behavior = people_behavior.groupby('gender')['user_id'].count()

gender_count_people_behavior.plot(kind='pie', autopct='%1.1f%%')
plt.title('Gender Distribution')
plt.ylabel('')  # Removes the default 'gender' y-label
plt.show()
```
