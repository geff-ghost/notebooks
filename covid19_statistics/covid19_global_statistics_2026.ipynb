```python
covid_df.columns
```

```python
total_cases = covid_df.total_cases.sum()
total_deaths = covid_df.total_deaths.sum()

print(f"The number of reported cases is {total_cases:.2f} and the number of reported deaths is {total_deaths:.2f}")
```
