## Numerical Computing with Python and Numpy

```python
w1, w2, w3 = 0.3, 0.2, 0.5
```

```python
kanto = [73, 67, 43]
johto = [91, 88, 64]
hoenn = [87, 134, 58]
sinnoh = [102, 43, 37]
unova = [69, 96, 70]

weights = [w1, w2, w3]
```

We can now write a function crop_yield to calculate the yield of apples (or any other crop) given the climate data and the respective weights.

```python
# This code loops through two lists simultaneously, pairing elements from region and weights together.
for x, w in zip(kanto, weights):
    print(x, w)
```

```python
def crop_yield(region, weights):
    results = 0
    for x, w in zip(region, weights):
        results += x*w
    return results
```

```python
crop_yield(kanto, weights)
```

## Operating on Numpy arrays  

We can now compute the dot product of the two vectors using the `np.dot` function.

```python
np.dot(kanto, weights)
```

We can achieve the same result with lower level operations supported by Numpy arrays:  

Performing an element-wise multiplication and calculating the sum of the resulting numbers.

```python
(kanto * weights).sum()
```

The `*` operator performs an element-wise multiplication of two arrays (assuming they have same size), and the sum method calculate the sum of numbers in an array.

```python
arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])

print(arr1*arr2)
```

**Benefits of Using Numpy arrays**  

- **Ease of Use**
- **Perfomance**

```python
# Python lists
arr1 = list(range(100000))
arr2 = list(range(100000, 200000))

# Numpy arrays
arr1_np = np.array(arr1)
arr2_np = np.array(arr2)
```

```python
%%time
result = 0
for x1, x2 in zip(arr1, arr2):
    result += x1*x2
result
```

```python
%%time
np.dot(arr1_np, arr2_np)
```

Using **np.dot** is 100 times faster than using a for loop  
- This makes Numpy especially useful while working with really large datasets with tens os thousands or millions of data points.

## Multi-dimesional Numpy arrays  

We can now go one step further, and represent the climate data for all the regions together using a single 2-dimensional Numpy array

```python
climate_data = np.array([
    [73, 67, 43],
    [91, 88, 64], 
    [102, 43, 37],
    [69, 96, 70]
])
```

```python
type(climate_data)
```

Numpy arrays can have any number of dimesions, and different lengths along each dimesion  

We can inspect the length along each dimesion using the `.shape` property of an array.

```python
# 2D array
climate_data.shape
```

```python
# 3D array
arr3 = np.array([
    [[11, 12, 13],
    [13, 14, 15]],
    [[15, 16, 17],
    [17, 18, 19.5]]
])
```

```python
arr3.shape
```

All the elements in a numpy array have the same datatype. You check the data type of an array using the `.dtype` property.

```python
weights.dtype
```

```python
climate_data.dtype
```

If an array contains even a single floating point number, all the other elements are also converted to floats.

We can now compute the predicted yields of apples in all the regions, using a single matrix multiplication between climate_data (a 5 by 3 matrix) and weights (a vector of length 3)  

We can use the `np.matmul` function from Numpy, or simply use the `@` operator to perform matrix multiplication.

```python
np.matmul(climate_data, weights)
```

```python
climate_data @ weights
```
